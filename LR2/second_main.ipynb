```python
import math
import matplotlib.pyplot as plt
import import_ipynb
from solver import iteration
```

```python
def r(x):
    return math.sin(18)


def p(x):
    return 0


def q(x):
    return 1 + math.cos(18) * x ** 2


def f(x):
    return -1
```

```python
borders = (
    (-1, 0),
    (1, 0),
)
func_coeff = (r, p, q, f)
eps = 0.001
result = iteration(func_coeff, borders, eps)
for x_vector, y_vector in result:
    print(max(y_vector))
    plt.plot(x_vector, y_vector)
    plt.grid()
plt.show()
```
